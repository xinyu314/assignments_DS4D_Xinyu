```python
import xml.etree.ElementTree as ET

import os
import numpy
```

```python
def namedic(file):
    tree = ET.parse(file)
    root = tree.getroot()
    
    list = []
    for meta in root.findall('meta'):
        name = meta.find('name').text
    list = [name]
    
    return(list)
    
#namedic('pta_1.xml')

rootdir = 'xml'
list = os.listdir(rootdir)
#df = pd.DataFrame(columns = ['Id','','date signed','composition','region','crossregional'])
#df

colum_1 = []  #source colum
colum_2 = []  #target colum
for i in range(0,len(list)):                # reading path of each agreement one by one 
    path = os.path.join(rootdir,list[i])
    name = namedic(path)[0]
    #print(name)
    colum = name.split('-')
    #print(colum)
    if len(colum) == 1:
        colum_2.append(' ')
    else:
        colum_2.append(colum[1])
    
#colum_2

```
